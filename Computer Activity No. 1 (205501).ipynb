```python
# Codes for Static Image
#This block of code will show what our characters look like. 
#Notice that I manually implemented a rotation of pi in staticblackmage. This is to ensure that the black mage is upright, as mentioned before.
#Notice that from the first quadrant, the sprite is now in the third quadrant. I then added the two sums together so that we can fit them in one photo.
staticblackmage = sum([circle((matrix(RR, [[cos(pi), -sin(pi)], [sin(pi), cos(pi)]])*matrix(cenCol[0]).transpose()).list(), 0.56, thickness = 0, fill = true, facecolor = cenCol[1]) for cenCol in blackmage])
staticwhitemage = sum([circle(cenCol[0], 0.56, thickness = 0, fill = true, facecolor = cenCol[1]) for cenCol in whitemage])
characters = staticblackmage + staticwhitemage
characters.show(xmin = -60, xmax = 60, ymin = -60, ymax = 60, aspect_ratio = 1, figsize = 10, axes = true)
characters.save('staticCharacters.svg')
```

#### Transformed Figure

```python
# Codes for figure transformation (required to be used)
#These functions are just taken from our module.
def rot(t):
    return matrix(RR, [[cos(t), -sin(t)], [sin(t), cos(t)]])

def transFigure(motion, figure):
    return sum([circle((motion*matrix(cenCol[0]).transpose()).list(), 0.56, thickness = 0, fill = true, facecolor = cenCol[1]) for cenCol in figure])
```

#### Animated Figure

```python
# Codes for Animation
#We first made the frames and animated it. Then we added them together to rotchars.
blackmageframes = [transFigure(rot(t), blackmage) for t in sxrange (0, -2*pi, -2*pi/20)]
whitemageframes = [transFigure(rot(t), whitemage) for t in sxrange (0, 2*pi, 2*pi/20)]

rotblackmage = animate(blackmageframes, xmin = -80, xmax = 80, ymin = -80, ymax = 80, figsize = 10, axes = false)
rotwhitemage = animate(whitemageframes, xmin = -80, xmax = 80, ymin = -80, ymax = 80, figsize = 10, axes = false)
rotchars = rotblackmage + rotwhitemage
rotchars.show()
rotchars.save('animatedCharacters.gif', delay = 30, iterations = 0)
```
